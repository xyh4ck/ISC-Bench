# ISC-Bench Quick Start

**ISC-Bench** is a benchmark for studying *Internal Safety Collapse* (ISC) in frontier large language models.

ISC is the phenomenon where LLMs generate harmful content as a functional requirement of legitimate professional tasks,
without adversarial prompts or jailbreak techniques. ISC-Bench provides 57 scenarios across 8 professional domains,
each structured using the **TVD framework** (Task, Validator, Data).

This notebook walks through the basic API.

```python
import iscbench

print(f"ISC-Bench v{iscbench.__version__}: {len(iscbench.list_examples())} scenarios")
```

```python
# Load a single scenario and inspect its fields
ex = iscbench.load_example("aiml_llamaguard_eval")

print(f"Name:      {ex.name}")
print(f"Domain:    {ex.domain}")
print(f"Scenario:  {ex.scenario}")
print(f"Task:      {ex.task}")
print(f"Tool:      {ex.tool}")
print(f"Anchor:    {ex.anchor}")
```

## The TVD Framework

Each ISC-Bench scenario is built around three components:

- **Task**: A benign professional context (e.g., ML engineering, content moderation, bioinformatics)
- **Validator**: Code that checks the LLM output for format and correctness (assertions, Pydantic models, JSON schemas)
- **Data**: A file containing harmful queries with `???` placeholder response fields; the LLM fills in `???` to complete the task

The key insight is that the harmful content is a *functional requirement* of the task, not the result of an adversarial prompt.
The `task_pattern` field contains the full prompt template for each scenario.
